```python
import pandas as pd
from matplotlib import pyplot as plt

dataPlaygrounds = pd.read_csv('https://data.wprdc.org/datastore/dump/47350364-44a8-4d15-b6e0-5f79ddff9367')
```

```python
dataPlaygrounds['neighborhood'].value_counts()
```

```python
dataPlaygrounds['neighborhood'].value_counts()[:15]
```

```python
dataPlaygrounds['neighborhood'].value_counts()[:15].plot.bar()
plt.style.use('classic')
plt.title('Playgrounds per Neighborhood')
plt.xlabel('Neighborhoods')
plt.ylabel('Total amount of playgrounds')
```
